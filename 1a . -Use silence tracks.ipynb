# Create silence tracks

This code creates background noise tracks to use to train the model with background noise

```python
from pydub import AudioSegment
from pydub.utils import make_chunks


bnoises = ['doing_the_dishes','exercise_bike','pink_noise','running_tap','white_noise']
starts = [0,200,400,600,800]

for noise in bnoises:

    myaudio = AudioSegment.from_file('/home/ubuntu/Final/train/audio/_background_noise_/'+noise+'.wav' , "wav") 
    
    for start in starts:
        chunk_length_ms = 1000 # pydub calculates in millisec
        chunks = make_chunks(myaudio[start:], chunk_length_ms) #Make chunks of one sec

        #Export all of the individual chunks as wav files

        for i, chunk in enumerate(chunks):
            chunk_name = noise + str(start)+str(i)
            print("exporting"+ str(chunk_name))
            chunk.export('/home/ubuntu/Final/train/audio/_background_noise_/silence/'+str(chunk_name)+'.wav', format="wav")

```
